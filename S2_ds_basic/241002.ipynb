```python
import pandas as pd

# 데이터프레임 생성
data = [
    ["1,000", "1,100", '1,510'],
    ["1,410", "1,420", '1,790'],
    ["850", "900", '1,185'],
]
columns = ["03/02", "03/03", "03/04"]
df = pd.DataFrame(data=data, columns=columns)
df.info()
```

```python
# 쉼표를 제거하고 int로 변환하는 함수 정의 

x = "1,000"
int(x.replace(",", ""))
```

```python
# 쉼표를 제거하고 int로 변환하는 함수 정의 

x = "1,000"
int(x.replace(",", ""))

def rm_comma(x):
    return int(x.replace(",", ""))

y = "1,000"
print(rm_comma(y))
```

```python
df["03/02"] = df["03/02"].apply(rm_comma)
df["03/03"] = df["03/03"].apply(rm_comma)
df.info()
```

```python
from pandas import DataFrame

data = [
    [1416, 1416, 2994, 1755],
    [6.42, 17.63, 21.09, 13.93],
    [1.10, 1.49, 2.06, 1.88]
]

columns = ["2018/12", "2019/12", "2020/12", "2020/12(E)"]
index = ["DPS", "PER", "PBR"]

df = DataFrame(data=data, index=index, columns=columns)
df
```

```python
df.filter(regex='2020')
```

```python
# 정규표현식으로 활용해서 행과 열을 조회
df.filter(items=['PER'], axis=0)
```

```python
# 행을 조회할 건데, R이 들어간 것만 조회 
df.filter(regex='R', axis = 0)

# 행을 조회할건데, R로 끝나는 모든 패턴을 조회
df.filter(regex='R$', axis=0)
```

```python
import plotly.express as px # seborn과 개념적으로 유사

iris = px.data.iris()       # iris.head() =>data 확인
fig = px.scatter(iris, x = 'sepal_length', y = 'sepal_width', color = 'species')

fig.add_trace(
    go.Scatter(
        x = [4.5, 7.5],
        y = [4, 2],
        # mode = "lines",
        # line = go.scatter.Line(color = "red"),
        mode = "markers",
        marker = dict(size = 20, color = 'red'),
        showlegend = False
    )
)
fig.show()
```

## Hover 설정 =>layout

```python
df = px.data.gapminder().query("continent == 'Oceania'")
df.tail()
```

```python
fig = px.line(df, x = 'year', y = 'lifeExp', color = 'country')

# 레이아웃은 그래프 외부의 것을 변경하는 것
# 그래프 내부의 색상, 모양 등등 변경 : update_traces()
fig.update_traces(mode="markers+lines")

# Hover 설정
fig.update_layout(hovermode = 'y')
fig.show()
```

## 캔들 차트

```python
import yfinance as yf
import plotly.graph_objects as go

indo = yf.download("INDO", start = "2023-01-01", end = "2024-09-30")
indo = indo.reset_index()
indo
```

```python
fig = go.Figure(data=[go.Candlestick(x=indo['Date'],
                open=indo['Open'], high=indo['High'],
                low=indo['Low'], close=indo['Close'])
                     ])

fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()


# 참고 : https://plotly.com/python/candlestick-charts/
```

```python
fig = go.Figure(data=[go.Candlestick(x=indo['Date'],
                open=indo['Open'], high=indo['High'],
                low=indo['Low'], close=indo['Close'])
                     ])

# update_laytout
# 차트 제목, 축 레이블 설정
fig.update_layout(
    title='Candlestick Chart',              # 차트 제목 설정
    xaxis_title='Date',                      # x축 레이블 설정
    yaxis_title='Price',                     # y축 레이블 설정
    xaxis_rangeslider_visible=False,         # x축 범위 슬라이더 숨기기
    template='plotly_white')
    

# update_traces
# 캔들차트의 색상을 변경 
fig.update_traces(
    increasing=dict(line=dict(color='green')),  # 상승 캔들 색상 설정
    decreasing=dict(line=dict(color='red'))     # 하락 캔들 색상 설정
)
    
fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```

```python
fig = go.Figure(data=[go.Candlestick(x=indo['Date'],
                open=indo['Open'], high=indo['High'],
                low=indo['Low'], close=indo['Close'])
                     ])

# update_laytout
# 차트 제목, 축 레이블 설정
fig.update_layout(
    title='인도네시아 에너지 차트(2023-2024)',             
    xaxis_title='날짜',                   
    yaxis_title='주식가격(USD)',               
    hovermode = 'x',
    xaxis_rangeslider_visible = True,
    xaxis_rangeslider_thickness = 0.5,
    paper_bgcolor = 'black',
    plot_bgcolor = 'white')
    

# update_traces
# 캔들차트의 색상을 변경 
fig.update_traces(
    increasing_line_color='red',  # 상승 캔들 색상 설정
    decreasing_line_color='blue')     # 하락 캔들 색상 설정

    
fig.show()
```
