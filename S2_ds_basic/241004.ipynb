## Concat

```python
from pandas import DataFrame
import pandas as pd

# 첫번째 데이터프레임
data = {
    '종가': [113000, 111500],
    '거래량': [555850, 282163]
}
index = ["2019-06-21", "2019-06-20"]
df1 = DataFrame(data, index=index)
df1
```

```python

# 두번째 데이터프레임
data = {
    '종가': [110000, 483689],
    '거래량': [109000, 791946]
}
index = ["2019-06-19", "2019-06-18"]

df2 = DataFrame(data, index =index)
df2
```

```python
result = pd.concat([df1, df2])   # => 데이터를 상하로 이어주는 것 
result

# 주의 : 합치고 나서 해결하는 것 보다 사전에 정리해서 합치는 것이 좋음 
```

## join 

```python
data = [
    ["전기전자", "005930", "삼성전자", 74400],
    ["화학", "051910", "LG화학", 896000],
    ["서비스업", "035750", "카카오", 101500]
]

columns = ["업종", "종목코드", "종목명", "현재가"]
df1 = DataFrame(data=data, columns=columns)
df1 = df1.set_index("업종")
print(df1)

data = [
    ["은행", 2.92],
    ["보험", 0.37],
    ["화학", 0.06],
    ["전기전자", -2.43]
]

columns = ["항목", "등락률"]
df2 = DataFrame(data=data, columns=columns)
df2 = df2.set_index("항목")
print(df2)
```

```python
# merge = 기본값 : inner join
# join = 기본값 : left join

df1.join(other = df2, how='left')
```

```python
# 항상 데이터를 어디서 가져왔는지, 왜 그 기준으로 데이터를 합쳐야하는지 생각해야함 .
```

```python
data = [
    ["2017", "삼성", 500],
    ["2017", "LG", 300],    
    ["2017", "SK하이닉스", 200],
    ["2018", "삼성", 600],
    ["2018", "LG", 400],
    ["2018", "SK하이닉스", 300],    
]

columns = ["연도", "회사", "시가총액"]
df = DataFrame(data=data, columns=columns)
df
```

```python
# 연도별로 그룹바이 연산, 시가총액 평균 
```

```python
result = df.groupby("연도")["시가총액"].mean().to_frame()
result.columns = ['시가총액평균']
result
```

```python
df = df.join(result, on = "연도")
df
```

```python
# => 초기 세팅을 동일하게 만들어주는게 좋음. 
```

```python
## 시가총액 평균을 기준값으로 
# 새로운 컬럼 : 규모 /
# 대형주, 중소형주
# numpy.where
```

```python
import numpy as np

df['규모'] = np.where(df['시가총액'] >= df['시가총액평균'], "대형주", "중소형주")
df
```

```python
df.drop('시가총액평균', axis = 1) 
# axis = 1 => 열삭제, axis = 0 => 행삭제
```
